```python
''' Correlation between variables '''
# Load in dataframe
df = pd.read_csv("bank_additional_full_1_comma_separate.csv")
df
```

```python
# Calculate correlation among numeric variables
corr = df.corr()
corr
```

```python
# Plot the correlation heatmap
fig = plt.figure()
ax = fig.add_subplot()
cax = ax.matshow(corr, cmap = 'coolwarm', vmin = -1, vmax = 1)
fig.colorbar(cax)
ticks = np.arange(0, len(corr.columns), 1)
ax.set_xticks(ticks)
plt.xticks(rotation = 90)
ax.set_yticks(ticks)
ax.set_xticklabels(corr.columns)
ax.set_yticklabels(corr.columns)
plt.show()
```

```python
# Check which two variables have strong correlation (corr > 0.5)
# Create col index
col_name = corr.columns.tolist()
strong_corr_pair = list()
for col_temp in corr.columns:
    col = col_name.index(col_temp)
    for row in range(0, col):
        if abs(corr.iloc[row, col]) > 0.5:
            if abs(corr.iloc[row, col]) == 1:
                pass
            else:
                strong_corr_pair.append([col_name[row], col_name[col]])
strong_corr_pair
```

```python
''' Handle missing values in the dataframe '''
# Check the number of missing values for each column
miss_value = list()
for col in df.columns:
    miss_value.append([col, df[col].dtype, len(df[df[col] == "unknown"])])
miss_value
```
